```python
import cv2
import numpy as np
from queue import Queue

# Load the image
image_path = '20 cells diameter theta maze.png'  # Replace with the actual path to your image
image = cv2.imread(image_path)# Replace with the actual path to your image
print("hello ")

try:
    image = cv2.imread(image_path)
    if image is None:
        raise ValueError("Failed to load the image.")
    grayscale = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
except Exception as e:
    print("Error: ", str(e))
    exit(1)

# Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(grayscale, (5, 5), 0)

# Apply adaptive thresholding to convert the image to binary
_, threshold = cv2.threshold(blurred, 127, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

# Apply Canny edge detection
edges = cv2.Canny(threshold, 50, 150)

# Find contours in the image
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Extract the outer contour of the maze
maze_contour = max(contours, key=cv2.contourArea)

# Create a blank image to draw the maze contour
maze_image = np.zeros_like(image)
cv2.drawContours(maze_image, [maze_contour], 0, (255, 255, 255), thickness=2)

# Convert the maze image to grayscale
maze_gray = cv2.cvtColor(maze_image, cv2.COLOR_BGR2GRAY)

# Find the start and end points of the maze
_, maze_binary = cv2.threshold(maze_gray, 127, 255, cv2.THRESH_BINARY)
maze_binary = cv2.bitwise_not(maze_binary)
start_point = np.argwhere(maze_binary == 0)[0]
end_point = np.argwhere(maze_binary == 0)[-1]

# Maze representation using a graph
maze_graph = {}
rows, cols = maze_gray.shape
for i in range(rows):
    for j in range(cols):
        if maze_gray[i, j] == 255:
            continue
        neighbors = []
        if i > 0 and maze_gray[i - 1, j] == 0:
            neighbors.append((i - 1, j))
        if i < rows - 1 and maze_gray[i + 1, j] == 0:
            neighbors.append((i + 1, j))
        if j > 0 and maze_gray[i, j - 1] == 0:
            neighbors.append((i, j - 1))
        if j < cols - 1 and maze_gray[i, j + 1] == 0:
            neighbors.append((i, j + 1))
        maze_graph[(i, j)] = neighbors

# Pathfinding algorithm (BFS)
def bfs(graph, start, end):
    visited = set()
    queue = Queue()
    queue.put([start])

    while not queue.empty():
        path = queue.get()
        node = path[-1]
        if node == end:
            return path
        if node not in visited:
            for neighbor in graph[node]:
                new_path = list(path)
                new_path.append(neighbor)
                queue.put(new_path)
            visited.add(node)

# Find the solution path
solution_path = bfs(maze_graph, start_point, end_point)



```

```python
import cv2
import numpy as np
import time

filename = '20 cells diameter theta maze'
img = cv2.imread(filename+'.png')

# Binary conversion
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV) 

# Contours
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
dc = cv2.drawContours(thresh, contours, 0, (255, 255, 255), 5)

dc = cv2.drawContours(dc, contours, 1, (0,0,0) , 5)
ret, thresh = cv2.threshold(dc, 240, 255, cv2.THRESH_BINARY)

ke = 19
kernel = np.ones((ke, ke), np.uint8)

# Dilate
dilation = cv2.dilate(thresh, kernel, iterations=1)

# Erosion
erosion = cv2.erode(dilation, kernel, iterations=1)

# Find differences between two images
diff = cv2.absdiff(dilation, erosion)

# splitting the channels of maze
b, g, r = cv2.split(img)
mask_inv = cv2.bitwise_not(diff)

# masking out the green and red colour from the solved path
r = cv2.bitwise_and(r, r, mask=mask_inv)
b = cv2.bitwise_and(b, b, mask=mask_inv)

res = cv2.merge((b, g, r))

# Find start and end points of maze
start_point = None
end_point = None
for i in range(res.shape[0]):
    if res[i,0,0] == 255:
        start_point = (i, 0)
        break
for i in range(res.shape[0]-1, -1, -1):
    if res[i,-1,0] == 255:
        end_point = (i, res.shape[1]-1)
        break

# Define function for depth-first search algorithm
def dfs(maze, start, end):
    stack = [(start, [start])]
    visited = set()
    while stack:
        (x,y), path = stack.pop()
        if (x,y) == end:
            return path
        if (x,y) not in visited:
            visited.add((x,y))
            if x > 0 and maze[x-1,y] == 255 and (x-1,y) not in visited:
                stack.append(((x-1,y), path + [(x-1,y)]))
            if x < maze.shape[0]-1 and maze[x+1,y] == 255 and (x+1,y) not in visited:
                stack.append(((x+1,y), path + [(x+1,y)]))
            if y > 0 and maze[x,y-1] == 255 and (x,y-1) not in visited:
                stack.append(((x,y-1), path + [(x,y-1)]))
            if y < maze.shape[1]-1 and maze[x,y+1] == 255 and (x,y+1) not in visited:
                stack.append(((x,y+1), path + [(x,y+1)]))
    # Return None if no solution path is found
    return None


# Solve maze using depth-first search algorithm and record time
# Solve maze using depth-first search algorithm and record time
start_time = time.time()
solution_path = dfs(r, start_point, end_point)
end_time = time.time()
elapsed_time = end_time - start_time

# Check if solution path was found
if solution_path is not None:
    # Draw solution path on maze
    for point in solution_path:
        r[point] = 255
        g[point] = 0
        b[point] = 0
else:
    print("No solution path found")

# Merge color channels and display solution maze with time taken to solve
res = cv2.merge((b, g, r))
cv2.putText(res, "Time taken: {:.2f} seconds".format(elapsed_time), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 255, 255), 2, cv2.LINE_AA)
cv2.imshow('Solved Maze', res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
